```python
# Con ayuda de una muestra de 506 comunidades en la zona de Boston, se estima un modelo que relaciona
# el precio medio de las viviendas (price) en una comunidad con diversas características de la misma: nox
# es la cantidad de óxido de nitrógeno en el aire, dada en partes por millón; dist es la distancia ponderada
# de la comunidad a cinco centros de trabajo, dada en millas; rooms es la cantidad promedio de habitaciones
# en las viviendas de la comunidad, y stratio es el cociente promedio estudiantes-profesores en las escuelas
# de la comunidad. El modelo poblacional es

import wooldridge as woo
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# woo.dataWoo('hprice2', description=True)
hprice2 = woo.dataWoo('hprice2')
hprice2

reg = smf.ols(
    formula='np.log(price) ~ np.log(nox)+np.log(dist)+rooms+I(rooms**2)+stratio',
    data=hprice2)
results = reg.fit()
print(f'results.summary(): \n{results.summary()}\n')
#predicciones con rooms = 4-8, todas las demás medias muestrales:
nox_mean = np.mean(hprice2['nox'])
dist_mean = np.mean(hprice2['dist'])
stratio_mean = np.mean(hprice2['stratio'])
X = pd.DataFrame({'rooms': np.linspace(4, 8, num=5),
                  'nox': nox_mean,
                  'dist': dist_mean,
                  'stratio': stratio_mean})
print(f'X: \n{X}\n')

# Calculamos el intervalo de confianza al 95%:
lpr_PICI = results.get_prediction(X).summary_frame(alpha=0.05)
print(lpr_PICI)
lpr_CI = lpr_PICI[['mean', 'mean_ci_lower', 'mean_ci_upper']]
print(f'lpr_CI: \n{lpr_CI}\n')

# plot:
plt.plot(X['rooms'], lpr_CI['mean'], color='blue',
         linestyle='-', label='')
plt.plot(X['rooms'], lpr_CI['mean_ci_upper'], color='red',
         linestyle='--', label='upper CI')
plt.plot(X['rooms'], lpr_CI['mean_ci_lower'], color='red',
         linestyle='--', label='lower CI')
plt.ylabel('lprice')
plt.xlabel('rooms')
plt.legend()
plt.savefig('Effects-Manual.pdf')
```
